# Scatter plot for Fees vs Discount_on_Fees

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Fees'], df['Discount_on_Fees'])
plt.xlabel('Fees')
plt.ylabel('Discount on Fees')
plt.title('Fees vs Discount on Fees')
plt.show()
```

# Scaling Data

```python
# Initialize StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler  # Import the StandardScaler

# Instantiate the scaler
scaler = StandardScaler()

# Create a scaled copy of the dataframe
df_scaled = df.copy()
df_scaled[['Fees', 'Discount_on_Fees']] = scaler.fit_transform(df[['Fees', 'Discount_on_Fees']])

# Display the first few rows of the scaled dataframe
print(df_scaled.head())
```
